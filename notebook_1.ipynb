```python
import pandas as pd;
import numpy as np;
```

```python
df_1 = pd.read_csv('fear_greed_index.csv')
df_2 = pd.read_csv('historical_data.csv')
```

```python
print(df_1.info())
print(df_2.info())
print(df_1.isnull().sum().head(10)) ## No null values
```

```python

df_1.columns = df_1.columns.str.lower().str.replace(' ', '_').str.strip()
df_2.columns = df_2.columns.str.lower().str.replace(' ', '_').str.strip()

#print(df_2.info())

df_2['correct_date']=pd.to_datetime(df_2['timestamp_ist'], format="%d-%m-%Y %H:%M")
df_1['correct_date_2']=pd.to_datetime(df_1['date'])

#print(df_1.dtypes)

#df_1['correct_date_2'].duplicated().sum()
#df_1['date'].duplicated().sum()


# Creating a column consiting only date format 
df_2['only_date_format']= df_2['correct_date'].dt.date

#print(df_2['only_date_format'])
#print(df_2['correct_date'])
#print(df_1['correct_date_2'])

# Changing the name of column for merging if needed

#df_2.rename(columns={"only_date_format": "only_date_format_common"}, inplace=True)
df_1.rename(columns={"correct_date_2": "only_date_format"}, inplace=True)
#print(df_2['only_date_format_common'])
#print(df_1['only_date_format_common'])

df_2['only_date_format']=pd.to_datetime(df_2['only_date_format'], format="%d-%m-%Y")
df_1.columns = df_1.columns.str.strip().str.lower()
df_1 = df_1.loc[:, ~df_1.columns.duplicated()]
#df_2['only_date_format_common']=pd.to_datetime(df_2['only_date_format'])
print(df_2['only_date_format'].dtypes)
print(df_2.columns)
print(df_2['only_date_format'])
print(df_1['only_date_format'].dtypes)
print(df_1.columns)
print(df_1['only_date_format'])
```

```python
## Insight 1 = Percentage of days falling on each classification

print(df_1['date'].count()) # The total number of days since there is no duplicated value

# Creating a function which will give percentage of days for each classification
def percentage(classification_type):
    days = df_1[df_1['classification']==classification_type]
    print_days = days['only_date_format'].count()
    #percentage_of_days = print_days/2644 * 100
    return print_days
# Put any desired filter in place of Extreme Fear for getting the percentage of days in that classification
percentage('Fear')

#Outcome = This will give us insight of getting the percentage of days in classification of day in trade market




```

```python
## Insight 2 = Distribution of Buy and Sell Trade 

def number_of_trades(call_of_trader):
    filter_1 =df_2[df_2['side']==call_of_trader]
    count_of_trades = filter_1['order_id'].count()
    return count_of_trades
#Enter the trader call for getting the total number of trades made on the given call
number_of_trades('BUY')
```

```python
## Insight 3 = Distribution of profitability

#1. Profitable vs Loss Vs Neutral Trade Numbers

user_input = input("Enter your scenario: ")
if user_input == 'Profit':
    profit_trade_number =df_2[df_2['closed_pnl']>0]
    print(len(profit_trade_number['side']))
elif user_input =='Loss':
    loss_trade_numbers = df_2[df_2['closed_pnl']<0]
    print(len(loss_trade_numbers['side']))
elif user_input =='Neutral':
    neutral_trade_numbers = df_2[df_2['closed_pnl']==0]
    print(len(neutral_trade_numbers['side']))
else:
    print("Enter Valid calssification type")
    
## Here we can also add percentage calculator to show the percentage of profit, loss and neutral trade numbers


```

```python
#2. Average profit vs loss values

average_value = input("Enter Your Desired Classification(profit or Loss):")

if average_value == 'Profit':
    profit_trade_number =df_2[df_2['closed_pnl']>0]
    avg_profit = profit_trade_number['closed_pnl'].mean()
    print(avg_profit)
elif average_value =='Loss':
    loss_trade_numbers = df_2[df_2['closed_pnl']<0]
    avg_loss = loss_trade_numbers['closed_pnl'].mean()
    print(avg_loss)
else:
    print("Enter Valid calssification type")
```

```python
# Merging the df_1 and df_2 on the common column only_date_format_common




print(df_2['only_date_format'].dtypes)
print(df_1['only_date_format'].dtypes)



df_3 = pd.merge(df_2, df_1, on='only_date_format', how='inner')

print(df_3.columns)
print(df_3[['side', 'only_date_format', 'classification']])

```

```python

## Insight 4 = Divergence in Buy and Sell Trades during diffrent market sentiments Checking

    

user_input_2 = input("Enter the sentiment you want to analyze (like= fear, greed etc. ...):")

def analysis_of_sentiment(user_input_2):
    type_of_trades = df_3[df_3['classification']==user_input_2]
    buy_trades = type_of_trades[type_of_trades['side']=='BUY']
    number_of_buy_trades = buy_trades['order_id'].count()
    sell_trades = type_of_trades[type_of_trades['side']=='SELL']
    number_of_sell_trades = sell_trades['order_id'].count()
    print("Buy Trades:", number_of_buy_trades)
    print("Sell Trades:", number_of_sell_trades)
    
analysis_of_sentiment(user_input_2)

# Key Findings
#1. Fear = buy and sell trades are almost equal ( 1k difference)
#2. Extreme Fear = same as above
#3. Greed = same as above
#4. Extreme Greed = buy trades are quite higher than sell trades which is obvious
#5. Neutral = again buy and sell trades are almost equal

```

```python
## Insight 5 = Profitability Distribution During Different Sentiments Of Market 

user_input_3 = input("Enter the sentiment you want to analyse:")

def profit_analysis(user_input_3):
    sentiment_type = df_3[df_3['classification']==user_input_3]
    profit_trades = sentiment_type[sentiment_type['closed_pnl']>0]
    number_of_profit_trades = profit_trades['order_id'].count()
    loss_trades = sentiment_type[sentiment_type['closed_pnl']<0]
    number_of_loss_trades = loss_trades['order_id'].count()
    neutral_trades = sentiment_type[sentiment_type['closed_pnl']==0]
    number_of_neutral_trades = neutral_trades['order_id'].count()
    print("Number Of Profit Trades During Fear Sentiment Are:", number_of_profit_trades)
    print("Number Of Loss Trades During Fear Sentiment Are:", number_of_loss_trades)
    print("Number Of Neutral Trades During Fear Sentiment Are:", number_of_neutral_trades)
    
profit_analysis(user_input_3)


# Key Findings
#1. During Fear Sentiment In Market = Neutral Trade numbers are high , profitable Trade number are also high, Loss trade numbers are quite lower
# Similarly for other sentiments >>> Mentioned in the Analysis Report in Detail
```

```python
df_3['crossed']=df_3['crossed'].astype(bool)
df_3['direction']=df_3['direction'].str.lower().str.replace(' ', '_').str.strip()
print(df_3['crossed'])
```

```python
## Insight 6 = Present vs Past Traders Behaviours on Direction

# Past ( crossed = True) it means the trade is closed amd Present Means the trade is active

user_input_4 = input("Past or Present Trends (write TRUE or FALSE respectively ) ?:").strip()
user_input_5 = input("Enter the direction you want to analyse(like = Open long, Close Long etc. ):").strip()

# Works for True/False typed by user (case-insensitive)
if user_input_4.lower() == "true":
    user_bool = True
elif user_input_4.lower() == "false":
    user_bool = False
else:
    raise ValueError("Invalid input! Please type True or False.")


def past_direction_analysis(user_bool, user_input_5):
    completed_trades = df_3[df_3['crossed'] == user_bool]
    direction_filter = completed_trades[completed_trades['direction']== user_input_5]
    total_size = direction_filter['size_usd'].sum()
    mean_size = direction_filter['size_usd'].mean()
    print("Total size in USD For your given Direction is:", total_size)
    print("Average size in USD For your given Direction is:", mean_size)
    
past_direction_analysis(user_bool, user_input_5)


# Key Features

# we can apply the filters respectively and can see the size in usd of the trades for desired directions
```
